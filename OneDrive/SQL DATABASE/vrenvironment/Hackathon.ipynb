```python
# Câu 8
df['month'] = df['InvoiceDate'].dt.month
countries = []
countries_app = []
df_2 = df[df['month'] < 4]
for i in df_2['Country']:
    if i in countries:
        countries_app.append(i)
        continue
    else:
        countries_app.append(i)
        countries.append(i)
appearance = []
for i in range(len(countries)):
    count = 0
    for item in countries_app:
        if item == countries[i]:
            count += 1
        else: continue
    appearance.append(count)
for i in countries:
    if countries.index(i) == appearance.index(max(appearance)):
        print('Quốc gia có số lượng đơn hàng bán ra nhiều nhất trong quý 1 là ' + i)
```

```python
# Câu 9
df_2 = df[df['month'] < 4]
for i in df_2['Country']:
    if i in countries:
        countries_app.append(i)
        continue
    else:
        countries_app.append(i)
        countries.append(i)
appearance = []
for i in range(len(countries)):
    count = 0
    for item in countries_app:
        if item == countries[i]:
            count += 1
        else: continue
    appearance.append(count)
top_5 = []
for item in range(len(countries)):
    for i in countries:
        if countries.index(i) == appearance.index(max(appearance)):
            top_5.append(i)
            if len(top_5) != 5:
                countries.remove(i)
                appearance.remove(max(appearance))
                continue
            else: 
                break
```

```python
df_2
```

```python
quy_1 = top_5[0:5]
stats = {}
quantity = []
for i in df_2['Quantity']:
    quantity.append(i)
for i in quy_1:
    total = 0    
    for item in countries_app:
        if item == i:
            for counter in range(len(countries_app)):
                if counter == countries_app.index(item):
                    total += quantity[counter]
                    break
    stats[i] = total
statis = dict(sorted(stats.items(), key = lambda x:x[1] , reverse=True))
fig, ax = plt.subplots(figsize=(20, 6))
plt.bar(statis.keys(), statis.values(), color = 'green')
plt.title('Top 5 quốc gia có số lượng sản phẩm cao nhất quý 1')
plt.xlabel('Các quốc gia')
plt.savefig('quy_1.png')
plt.show()
```
